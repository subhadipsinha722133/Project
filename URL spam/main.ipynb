```python
import pandas as pd
a=pd.read_csv("url_spam_classification.csv")
a
```

```python
a=a.iloc[:35000]
```

```python
a.duplicated().sum()
```

```python
a.drop_duplicates(inplace=True)
```

```python
a.isnull().sum()
```

```python
a.info()
```

```python
import re
def remove_html(text):
    p=re.compile("<.?>")
    return p.sub(r"",text)

a["url"]=a["url"].apply(remove_html)
```

```python
def remove_url(text):
    p=re.compile(r'http?//\S+|www\.S+')
    return p.sub(r'',text)
```

```python
a["url"]=a["url"].apply(remove_url)
```

```python
a["url"]=a["url"].str.lower()
```

```python
from nltk.corpus import stopwords

def remove_stopword(text):
    new_text=[]
    for i in text.split():
        if i in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(i)
    x=new_text[ : ]
    new_text.clear()
    return " ".join(x)

a["url"]=a["url"].apply(remove_stopword)
```

```python
import spacy
n=spacy.load("en_core_web_sm")
a["url"]=a["url"].astype(str).apply(n)

a["url"]=a["url"].apply(lambda d: " ".join([t.lemma_ for t in d]) )
```

```python
a
```

```python
x=a.iloc[:,:-1]
y=a.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer()
x_train=tfidf.fit_transform(x_train["url"]).toarray()
x_test=tfidf.transform(x_test["url"]).toarray()
```

```python
x_train.shape
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
# 1. Import the necessary components
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam # <-- Import the optimizer

# Your model definition remains the same
model = Sequential()
model.add(Dense(500, input_dim=31885, activation="sigmoid"))
model.add(Dense(500, activation="relu"))
model.add(Dense(500, activation="relu"))
model.add(Dense(1, activation="tanh"))

# 2. Instantiate the optimizer in the compile() method
# Instead of optimizer="Adam"
# Use       optimizer=Adam()
model.compile(optimizer=Adam(), loss="mse", metrics=["accuracy"])

# You can also customize the learning rate this way
# model.compile(optimizer=Adam(learning_rate=0.001), loss="mse")

model.summary()
```

```python
model.fit(x_train,y_train,epochs=4,validation_data=(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rr=RandomForestClassifier()
rr.fit(x_train,y_train)
rr.score(x_test,y_test)*100
```
