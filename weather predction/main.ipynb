```python
a.drop(columns=['Date', 'Location','WindGustDir'],inplace=True)
```

```python
flot_data=a.select_dtypes(include="float64")

from sklearn.impute import SimpleImputer

sim=SimpleImputer(strategy="mean")
# data=sim.fit_transform(flot_data)
for i in flot_data.keys():
    a[i]=sim.fit_transform(flot_data[[i]])

```

```python
ob_data=a.select_dtypes(include="object")

from sklearn.impute import SimpleImputer
sim=SimpleImputer(strategy="most_frequent")

for i in ob_data.keys():
    a[[i]]=sim.fit_transform(ob_data[[i]])
```

```python
ob=a.select_dtypes(include="object")

from sklearn.preprocessing import LabelEncoder

la=LabelEncoder()

for i in ob.keys():
    a[i]=la.fit_transform(ob[[i]])
```

```python
x=a.iloc[:,:-3]

y=a.iloc[:,-2]
z=a.iloc[:,-1]
```

```python
yz=a.iloc[:,-2:]
yz
```

```python
from imblearn.over_sampling import RandomOverSampler

ros=RandomOverSampler()
x,z=ros.fit_resample(x,z)

```

```python


from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,z,random_state=42,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

st=StandardScaler()
x_train=st.fit_transform(x_train)
x_test=st.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

Dtc=DecisionTreeClassifier()
Dtc.fit(x_train,y_train)
Dtc.score(x_test,y_test)*100
```

```python
y_pred=Dtc.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfr=RandomForestClassifier(n_estimators=50,max_depth=50,criterion="entropy")
rfr.fit(x_train,y_train)
rfr.score(x_test,y_test)*100
```

```python
t_pred=rfr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,t_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
